```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
original = pd.read_csv("Top500_Indian_Cities.csv")
```

Multiplying all population by 3

```python
original.population_total = 3*original.population_total 
```

```python
state_wise = original.groupby(['state_name']).sum()
```

```python
highest_pop = state_wise.sort_values(by=['population_total'], ascending = False).head(10)
```

```python
highest_pop = highest_pop.rename_axis('state_name').reset_index()
```

```python
x = highest_pop.state_name
```

```python
y = highest_pop.population_total
```

```python
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print ("Current size:", fig_size)
```

```python
fig_size[0] = 8
fig_size[1] = 4
plt.rcParams["figure.figsize"] = fig_size
```

```python
plt.barh(x,y, color=['yellow','orange'], height = 0.8)
plt.title('Highest Population')
plt.xlabel('Population (in crores)')
plt.gca().invert_yaxis()
for i, v in enumerate(y):
    plt.text(v - 15000000, i + 0.2 , str(v), color='black', fontweight='regular')
plt.savefig('Highest Population.png', pad_inches = 0.4)    
```

```python
state_wise['literacy_rate'] = (state_wise.literates_total/state_wise.population_total)*100
```

As Total Population was multiplied by a factor of 3, the literacy rate will be one third of the original values

```python
literacy = state_wise.sort_values(by=['literacy_rate'], ascending = False).head(10)
```

```python
literacy = literacy.rename_axis('state_name').reset_index()
```

```python
x = literacy.state_name
```

```python
y = literacy.literacy_rate
```

```python
y = round(y,2)
```

```python
fig_size[0] = 8
fig_size[1] = 6
plt.rcParams["figure.figsize"] = fig_size
```

```python
plt.barh(x,y, color=['yellow','orange'], height = 0.8)
plt.title('Highest Literacy')
plt.xlabel('Literacy Rate (%)')
plt.gca().invert_yaxis()
for i, v in enumerate(y):
    plt.text(v - 3, i + 0.2 , str(v), color='black', fontweight='regular')
plt.savefig('Highest Literacy.png', pad_inches = 0.4)    
```

```python
state_wise['sex_ratio'] = (state_wise.population_female/state_wise.population_male)*1000
```

```python
sex_ratio = state_wise.sort_values(by=['sex_ratio'], ascending = False).head(10)
```

```python
sex_ratio = sex_ratio.rename_axis('state_name').reset_index()
```

```python
sex_ratio
```

```python
x = sex_ratio.state_name
```

```python
y = sex_ratio.sex_ratio
```

```python
y = round(y)
```

```python
plt.barh(x,y, color=['yellow','orange'], height = 0.8)
plt.title('Sex Ratio')
plt.xlabel('Women per 1000 Men')
plt.gca().invert_yaxis()
for i, v in enumerate(y):
    plt.text(v - 100, i + 0.2 , str(v), color='black', fontweight='regular')
plt.savefig('Highest Sex Ratio.png', pad_inches = 0.4)    
```

```python
state_wise['graduate_percentage'] = (state_wise.total_graduates/state_wise.population_total)*100
```

```python
graduate = state_wise.sort_values(by=['graduate_percentage'], ascending = False).head(10)
```

```python
graduate = graduate.rename_axis('state_name').reset_index()
```

```python
graduate
```

```python
x = graduate.state_name
```

```python
y = graduate.graduate_percentage
```

```python
y = round(y,2)
```

```python
plt.barh(x,y, color=['green','lime'], height = 0.8)
plt.title('Graduate Ratio')
plt.xlabel('Percentage of Graduates')
plt.gca().invert_yaxis()
for i, v in enumerate(y):
    plt.text(v - 0.75, i + 0.2 , str(v), color='black', fontweight='regular')
plt.savefig('Highest Graduate Ratio.png', pad_inches = 0.4)    
```
